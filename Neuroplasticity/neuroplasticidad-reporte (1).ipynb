```python

#!/usr/bin/env python3
"""
Plastic-LSTM v3.3 - Neuroplasticity with Individual Neuron Tracking
==================================================================
Enhanced version with:
- Vectorized operations for GPU efficiency
- Memory-efficient sampling
- Phase B data injection
- Auto-saving plots for Kaggle
- Adaptive threshold regulation
"""

import os, random, math, time, sys, collections, itertools
import numpy as np
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, Subset
from collections import defaultdict, deque
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from datetime import datetime

# Deterministic behavior - DISABLED for CUDA compatibility
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# Dark theme
sns.set_style("darkgrid", {"axes.facecolor": "#1c1c1c", "grid.color": "#444"})
plt.rcParams.update({
    "figure.facecolor": "#1c1c1c",
    "savefig.facecolor": "#1c1c1c",
    "text.color": "white",
    "axes.labelcolor": "white",
    "axes.edgecolor": "#aaa"
})

# Create output directories
os.makedirs('/kaggle/working/plots', exist_ok=True)
os.makedirs('/kaggle/working/checkpoints', exist_ok=True)

class Config:
    # Data
    sequence_length = 128
    batch_size = 32  # Smaller for faster testing
    max_data_mb = 50
    
    # Training - Quick test
    learning_rate = 5e-4
    num_epochs = 5
    eval_every_steps = 100
    
    # Model
    embedding_dim = 256
    lstm_hidden_size = 512
    num_lstm_layers = 2
    vocab_size = None  # Set dynamically
    
    # Plasticity - MORE AGGRESSIVE with ADAPTIVE thresholds
    max_neurons = 1024
    initial_active = 256
    plasticity_interval = 25
    maturation_time = 500
    
    # ADAPTIVE THRESHOLDS
    birth_threshold_loss = 3.6
    death_threshold_contrib = 0.001
    birth_burst_rate = 0.05
    
    # Adaptive regulation parameters
    adaptive_thresholds = True
    target_birth_rate = 0.02
    target_death_rate = 0.01
    threshold_adapt_rate = 0.1
    
    # Hebbian parameters
    hebb_eta = 0.01
    hebb_decay = 0.99
    hebb_norm_clip = 5.0
    
    # Memory efficiency
    track_neurons = True
    neuron_history_size = 100  # Reduced from 1000
    sampling_rate = 0.1  # Only track 10% of activations
    metric_log_interval = 200
    
    test_prompts = [
        "The quick brown fox",
        "Once upon a time",
        "import torch",
        "def forward(self",
        "The meaning of life",
    ]
    
    # Experiment phases
    phase_b_start = 5000
    phase_b_data_mb = 20  # Additional data for phase B
    
    # Ablation flags
    disable_hebbian = False
    disable_death = False
    
    # Replay
    replay_size = 100  # Reduced for memory
    
    # Paths
    data_path = '/kaggle/input/aaaaaad/openwebtext_2GB.txt'
    checkpoints_dir = '/kaggle/working/checkpoints'
    plots_dir = '/kaggle/working/plots'

cfg = Config()

class CharDataset(Dataset):
    def __init__(self, text, seq_len, stoi=None, itos=None):
        if stoi is None:
            chars = sorted(set(text))
            self.stoi = {c: i for i, c in enumerate(chars)}
            self.itos = {i: c for c, i in self.stoi.items()}
        else:
            self.stoi, self.itos = stoi, itos
        
        self.data = [self.stoi.get(c, 0) for c in text]
        self.seq_len = seq_len
        
    def __len__(self):
        return max(1, (len(self.data) - 1) // self.seq_len)
    
    def __getitem__(self, idx):
        s = idx * self.seq_len
        chunk = self.data[s:s + self.seq_len + 1]
        if len(chunk) < self.seq_len + 1:
            chunk = chunk + [0] * (self.seq_len + 1 - len(chunk))
        return torch.tensor(chunk[:-1]), torch.tensor(chunk[1:])

def build_loaders(cfg, phase='A', existing_stoi=None, existing_itos=None):
    """Build data loaders with phase support"""
    print(f'📚 Loading text for Phase {phase}...')
    
    try:
        with open(cfg.data_path, 'r', encoding='utf-8', errors='ignore') as f:
            if phase == 'A':
                text = f.read(cfg.max_data_mb * 1024 * 1024)
            else:  # Phase B
                # Skip initial data and read new chunk
                f.seek(cfg.max_data_mb * 1024 * 1024)
                text = f.read(cfg.phase_b_data_mb * 1024 * 1024)
        print(f"Loaded {len(text)/1e6:.1f}M characters for Phase {phase}")
    except FileNotFoundError:
        print("Dataset not found, using synthetic data")
        if phase == 'A':
            text = "The quick brown fox jumps over the lazy dog. " * 100000
        else:
            text = "In a hole in the ground there lived a hobbit. " * 100000
    
    # Use existing vocabulary for phase B
    full = CharDataset(text, cfg.sequence_length, existing_stoi, existing_itos)
    
    if phase == 'A':
        cfg.vocab_size = len(full.itos)
        print(f"Vocabulary size: {cfg.vocab_size}")
    
    n = len(full)
    train_size = int(0.9 * n)
    train, val = torch.utils.data.random_split(
        full, [train_size, n - train_size],
        generator=torch.Generator().manual_seed(SEED + (1 if phase == 'B' else 0))
    )
    
    mk = lambda ds, shuf: DataLoader(
        ds, batch_size=cfg.batch_size, shuffle=shuf, 
        num_workers=0, drop_last=True
    )
    
    return mk(train, True), mk(val, False), full.stoi, full.itos, full

class ReplayBuffer:
    def __init__(self, max_batches):
        self.buf = deque(maxlen=max_batches)
    
    def push(self, x, y):
        self.buf.append((x.cpu(), y.cpu()))
    
    def sample(self, n=1):
        if not self.buf:
            return None
        if n == 1:
            return random.choice(self.buf)
        return random.choices(self.buf, k=min(n, len(self.buf)))

class PlasticLSTM(nn.Module):
    def __init__(self, c):
        super().__init__()
        self.c = c
        
        # Core model
        self.embed = nn.Embedding(c.vocab_size, c.embedding_dim)
        self.lstm = nn.LSTM(
            c.embedding_dim, c.lstm_hidden_size, c.num_lstm_layers,
            batch_first=True, dropout=0.1
        )
        self.W = nn.Linear(c.lstm_hidden_size, c.max_neurons)
        self.out = nn.Linear(c.max_neurons, c.vocab_size)
        
        # Plasticity buffers
        self.register_buffer('mask', torch.zeros(c.max_neurons))
        self.mask[:c.initial_active] = 1
        self.register_buffer('age', torch.zeros(c.max_neurons))
        self.register_buffer('contrib', torch.zeros(c.max_neurons))
        self.register_buffer('h_fast', torch.zeros(
            c.max_neurons, c.lstm_hidden_size, dtype=torch.float16
        ))
        self.register_buffer('firing_count', torch.zeros(c.max_neurons))
        
        # Gradient-based contribution tracking
        self.register_buffer('grad_contrib', torch.zeros(c.max_neurons))
        self.register_buffer('output_impact', torch.zeros(c.max_neurons))
        
        # Tracking
        self.life_events = []
        self.step = 0
        self._cache = (None, None)
        
        # Advanced metrics tracking (memory efficient)
        self.age_hist = deque(maxlen=50)
        self.contrib_hist = deque(maxlen=50)
        self.fw_norm_hist = deque(maxlen=50)
        self.act_dist_hist = deque(maxlen=50)
        
        # Neuron activation history (with sampling)
        if c.track_neurons:
            self.neuron_history = defaultdict(lambda: deque(maxlen=c.neuron_history_size))
            self.prompt_activations = defaultdict(list)
        
        # Adaptive threshold tracking
        self.current_birth_threshold = c.birth_threshold_loss
        self.current_death_threshold = c.death_threshold_contrib
        self.recent_births = deque(maxlen=100)
        self.recent_deaths = deque(maxlen=100)
        
        # Phase tracking
        self.current_phase = 'A'
        
        # Initialize weights properly
        with torch.no_grad():
            nn.init.xavier_uniform_(self.W.weight)
            nn.init.zeros_(self.W.bias)
            nn.init.xavier_uniform_(self.out.weight)
    
    def forward(self, x):
        self.step += 1
        
        # Forward pass
        h, _ = self.lstm(self.embed(x))
        last = h[:, -1, :]
        
        # Combine slow and fast weights
        if self.c.disable_hebbian:
            W_eff = self.W.weight
        else:
            h_fast_normed = self.h_fast.to(self.W.weight.dtype)
            norms = h_fast_normed.norm(dim=1, keepdim=True)
            h_fast_normed = h_fast_normed / (norms.clamp(min=1e-8) / self.c.hebb_norm_clip).clamp(min=1.0)
            W_eff = self.W.weight + self.c.hebb_eta * h_fast_normed
        
        # Compute neuron activations
        pre_acts = F.linear(last, W_eff * self.mask.unsqueeze(1), self.W.bias)
        acts = torch.sigmoid(pre_acts)
        
        # Output
        logits = self.out(acts)
        
        # Cache for updates
        self._cache = (acts.detach(), last.detach())
        
        # Track gradient contribution
        if self.training and acts.requires_grad:
            def capture_grad(grad):
                if grad is not None:
                    self.grad_contrib = 0.95 * self.grad_contrib + 0.05 * grad.abs().mean(0)
            
            if acts.grad_fn is not None:
                acts.register_hook(capture_grad)
        
        # Track metrics (with sampling for memory efficiency)
        if self.training:
            with torch.no_grad():
                # Update firing rate
                firing_update = (acts > 0.5).float().mean(0) * self.mask
                self.firing_count += firing_update
                
                # Output impact
                if logits.shape[0] > 0:
                    neuron_variance = acts.std(0) * self.mask
                    self.output_impact = 0.9 * self.output_impact + 0.1 * neuron_variance
                
                # Sample neuron activations (memory efficient)
                if self.c.track_neurons and random.random() < self.c.sampling_rate:
                    for i in range(self.c.max_neurons):
                        if self.mask[i] > 0:
                            self.neuron_history[i].append(acts[:, i].mean().item())
            
            # Log detailed metrics periodically
            if self.step % self.c.metric_log_interval == 0:
                self._log_detailed_metrics(acts)
        
        return logits
    
    def _log_detailed_metrics(self, acts):
        """Log detailed metrics for analysis (memory efficient)"""
        self.age_hist.append(self.age.cpu().clone())
        self.contrib_hist.append(self.contrib.cpu().clone())
        
        fw_norm = self.h_fast.float().norm(dim=1)
        self.fw_norm_hist.append(fw_norm.cpu())
        
        act_sample = acts.detach().cpu().flatten()[:1000]
        self.act_dist_hist.append({
            'mean': act_sample.mean().item(),
            'std': act_sample.std().item(),
            'max': act_sample.max().item(),
            'data': act_sample.numpy()
        })
    
    @torch.no_grad()
    def hebbian_update(self):
        """Update fast weights using Hebbian rule"""
        if self.c.disable_hebbian:
            return
            
        acts, last = self._cache
        if acts is None:
            return
        
        # Center activations for stability
        acts_centered = acts - acts.mean(1, keepdim=True)
        last_centered = last - last.mean(1, keepdim=True)
        
        outer = acts_centered.unsqueeze(2) * last_centered.unsqueeze(1)
        dw = outer.mean(0).to(self.h_fast.dtype)
        
        # Update with decay
        mask_expanded = self.mask.unsqueeze(1).to(self.h_fast.dtype)
        self.h_fast = self.h_fast * self.c.hebb_decay * mask_expanded + dw * mask_expanded
        
        # Clip norms
        norms = self.h_fast.norm(dim=1, keepdim=True)
        self.h_fast = self.h_fast / (norms / self.c.hebb_norm_clip).clamp(min=1.0)
    
    @torch.no_grad()
    def structural_update(self, loss):
        """VECTORIZED birth and death of neurons"""
        # Update age
        self.age += self.mask
        
        # Update contribution metrics
        acts, _ = self._cache
        if acts is not None:
            act_contrib = acts.mean(0) * self.mask
            combined_contrib = (
                0.4 * act_contrib +
                0.3 * self.grad_contrib +
                0.3 * self.output_impact
            )
            self.contrib = 0.9 * self.contrib + 0.1 * combined_contrib
        
        births_this_update = 0
        deaths_this_update = 0
        
        # VECTORIZED DEATH
        if not self.c.disable_death:
            death_threshold = self.current_death_threshold if self.c.adaptive_thresholds else self.c.death_threshold_contrib
            
            # Find dying neurons
            dying_mask = (self.age > self.c.maturation_time) & (self.contrib < death_threshold)
            death_idx = dying_mask.nonzero(as_tuple=True)[0]
            
            # Limit deaths
            max_deaths = int(self.mask.sum() * 0.05)
            if len(death_idx) > max_deaths:
                contrib_values = self.contrib[death_idx]
                _, sorted_indices = contrib_values.sort()
                death_idx = death_idx[sorted_indices[:max_deaths]]
            
            if len(death_idx) > 0:
                deaths_this_update = len(death_idx)
                
                # VECTORIZED OPERATIONS
                self.mask[death_idx] = 0
                self.age[death_idx] = 0
                self.contrib[death_idx] = 0
                self.grad_contrib[death_idx] = 0
                self.output_impact[death_idx] = 0
                self.W.weight.data[death_idx] = 0
                self.W.bias.data[death_idx] = 0
                self.out.weight.data[:, death_idx] = 0
                self.h_fast[death_idx] = 0
                self.firing_count[death_idx] = 0
                
                # Log events
                for idx in death_idx:
                    self.life_events.append({
                        'type': 'death',
                        'step': self.step,
                        'id': idx.item(),
                        'phase': self.current_phase
                    })
                
                print(f"  💀 Step {self.step}: Killed {deaths_this_update} neurons")
        
        # VECTORIZED BIRTH
        birth_threshold = self.current_birth_threshold if self.c.adaptive_thresholds else self.c.birth_threshold_loss
        active_ratio = self.mask.sum() / self.c.max_neurons
        population_pressure = max(0, 0.3 - active_ratio) * 5
        
        if loss > birth_threshold - population_pressure:
            free_mask = (self.mask == 0)
            free_idx = free_mask.nonzero(as_tuple=True)[0]
            
            if len(free_idx) > 0:
                loss_factor = min((loss - birth_threshold + population_pressure) / 2.0, 1.0)
                num_births = max(1, int(len(free_idx) * self.c.birth_burst_rate * loss_factor))
                num_births = min(num_births, len(free_idx))
                
                birth_idx = free_idx[torch.randperm(len(free_idx))[:num_births]]
                births_this_update = len(birth_idx)
                
                # VECTORIZED OPERATIONS
                self.mask[birth_idx] = 1
                self.age[birth_idx] = 0
                self.contrib[birth_idx] = 0.1
                self.grad_contrib[birth_idx] = 0
                self.output_impact[birth_idx] = 0
                self.h_fast[birth_idx] = 0
                self.firing_count[birth_idx] = 0
                
                # Initialize weights (still need loop for init functions)
                for idx in birth_idx:
                    nn.init.xavier_uniform_(self.W.weight.data[idx:idx+1])
                    nn.init.normal_(self.out.weight.data[:, idx], 0, 0.02)
                
                self.W.bias.data[birth_idx] = 0
                
                # Log events
                for idx in birth_idx:
                    self.life_events.append({
                        'type': 'birth',
                        'step': self.step,
                        'id': idx.item(),
                        'loss': loss,
                        'phase': self.current_phase
                    })
                
                print(f"  🌟 Step {self.step}: Born {births_this_update} neurons (loss={loss:.3f})")
        
        # Update adaptive thresholds
        if self.c.adaptive_thresholds and self.step % 100 == 0:
            self._update_adaptive_thresholds(births_this_update, deaths_this_update)
        
        self.recent_births.append(births_this_update)
        self.recent_deaths.append(deaths_this_update)
    
    def _update_adaptive_thresholds(self, births, deaths):
        """Adaptively adjust thresholds"""
        total_births = sum(self.recent_births)
        total_deaths = sum(self.recent_deaths)
        updates = len(self.recent_births)
        
        if updates > 10:
            active_neurons = self.mask.sum().item()
            birth_rate = total_births / (updates * max(1, self.c.max_neurons - active_neurons))
            death_rate = total_deaths / (updates * max(1, active_neurons))
            
            # Adjust thresholds
            if birth_rate < self.c.target_birth_rate:
                self.current_birth_threshold *= (1 - self.c.threshold_adapt_rate)
            elif birth_rate > self.c.target_birth_rate * 2:
                self.current_birth_threshold *= (1 + self.c.threshold_adapt_rate)
            
            if death_rate < self.c.target_death_rate and active_neurons > self.c.initial_active:
                self.current_death_threshold *= (1 + self.c.threshold_adapt_rate)
            elif death_rate > self.c.target_death_rate * 2:
                self.current_death_threshold *= (1 - self.c.threshold_adapt_rate)
            
            # Clamp
            self.current_birth_threshold = torch.clamp(
                torch.tensor(self.current_birth_threshold), 2.5, 5.0
            ).item()
            self.current_death_threshold = torch.clamp(
                torch.tensor(self.current_death_threshold), 1e-4, 1e-2
            ).item()
    
    def set_phase(self, phase):
        """Change experimental phase"""
        self.current_phase = phase
        print(f"\n🔬 Switched to Phase {phase}")
    
    @torch.no_grad()
    def test_prompt(self, prompt, stoi, device, num_runs=5):
        """Test prompt and track activations"""
        self.eval()
        
        tokens = [stoi.get(c, 0) for c in prompt]
        if len(tokens) < self.c.sequence_length:
            tokens = tokens + [0] * (self.c.sequence_length - len(tokens))
        else:
            tokens = tokens[:self.c.sequence_length]
        
        x = torch.tensor([tokens]).to(device)
        
        all_acts = []
        for run in range(num_runs):
            logits = self(x)
            acts, _ = self._cache
            all_acts.append(acts[0].cpu().numpy())
        
        acts_mean = np.mean(all_acts, axis=0)
        acts_std = np.std(all_acts, axis=0)
        
        self.prompt_activations[prompt].append({
            'step': self.step,
            'mean': acts_mean,
            'std': acts_std,
            'active_neurons': (self.mask > 0).cpu().numpy(),
            'phase': self.current_phase
        })
        
        self.train()
        return acts_mean, acts_std
    
    def get_neuron_info(self, neuron_id):
        """Get detailed neuron info"""
        info = {
            'id': neuron_id,
            'active': bool(self.mask[neuron_id] > 0),
            'age': int(self.age[neuron_id]),
            'contribution': float(self.contrib[neuron_id]),
            'grad_contribution': float(self.grad_contrib[neuron_id]),
            'output_impact': float(self.output_impact[neuron_id]),
            'firing_rate': float(self.firing_count[neuron_id] / max(1, self.step)),
            'fw_norm': float(self.h_fast[neuron_id].float().norm()),
            'history': list(self.neuron_history.get(neuron_id, [])),
            'birth_death': []
        }
        
        for event in self.life_events:
            if event['id'] == neuron_id:
                info['birth_death'].append(event)
        
        return info

class NeuronAnalyzer:
    def __init__(self, model, cfg, val_loader, itos):
        self.model = model
        self.cfg = cfg
        self.val_loader = val_loader
        self.itos = itos
        self.history = defaultdict(list)
    
    @torch.no_grad()
    def evaluate(self, loader, device):
        self.model.eval()
        total_loss = 0
        n = 0
        
        for x, y in loader:
            x, y = x.to(device), y.to(device)
            logits = self.model(x)
            loss = F.cross_entropy(logits, y[:, -1])
            total_loss += loss.item() * x.size(0)
            n += x.size(0)
        
        avg_loss = total_loss / n
        ppl = math.exp(avg_loss)
        return avg_loss, ppl
    
    def log_metrics(self, step, loss, ppl, active_neurons, phase='A'):
        self.history['step'].append(step)
        self.history['loss'].append(loss)
        self.history['ppl'].append(ppl)
        self.history['active'].append(active_neurons)
        self.history['phase'].append(phase)
        
        if self.cfg.adaptive_thresholds:
            self.history['birth_threshold'].append(self.model.current_birth_threshold)
            self.history['death_threshold'].append(self.model.current_death_threshold)
    
    def save_plot(self, fig, name):
        """Save plot to Kaggle working directory"""
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"{self.cfg.plots_dir}/{name}_{timestamp}.html"
        fig.write_html(filename)
        print(f"💾 Saved plot: {filename}")
        return filename
    
    def save_matplotlib_plot(self, name):
        """Save matplotlib plot"""
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"{self.cfg.plots_dir}/{name}_{timestamp}.png"
        plt.savefig(filename, dpi=150, bbox_inches='tight')
        print(f"💾 Saved plot: {filename}")
        return filename
    
    def plot_training_progress(self):
        """Plot training metrics with phase markers"""
        fig = make_subplots(
            rows=5, cols=1,
            subplot_titles=('Loss', 'Perplexity', 'Active Neurons', 'Birth/Death Events', 'Adaptive Thresholds'),
            vertical_spacing=0.08
        )
        
        # Add phase backgrounds
        if 'phase' in self.history:
            phase_changes = []
            for i in range(1, len(self.history['phase'])):
                if self.history['phase'][i] != self.history['phase'][i-1]:
                    phase_changes.append(self.history['step'][i])
            
            for step in phase_changes:
                for row in range(1, 6):
                    fig.add_vline(x=step, line_dash="dash", line_color="yellow", opacity=0.3,
                                annotation_text="Phase B", annotation_position="top", row=row, col=1)
        
        # Loss
        fig.add_trace(
            go.Scatter(x=self.history['step'], y=self.history['loss'], 
                      name='Loss', line=dict(color='cyan')),
            row=1, col=1
        )
        
        # Perplexity
        fig.add_trace(
            go.Scatter(x=self.history['step'], y=self.history['ppl'], 
                      name='PPL', line=dict(color='magenta')),
            row=2, col=1
        )
        
        # Active neurons
        fig.add_trace(
            go.Scatter(x=self.history['step'], y=self.history['active'], 
                      name='Active', line=dict(color='lime')),
            row=3, col=1
        )
        
        # Birth/Death events
        if self.model.life_events:
            births = [(e['step'], e.get('phase', 'A')) for e in self.model.life_events if e['type'] == 'birth']
            deaths = [(e['step'], e.get('phase', 'A')) for e in self.model.life_events if e['type'] == 'death']
            
            if births:
                birth_steps, birth_phases = zip(*births)
                colors = ['green' if p == 'A' else 'lightgreen' for p in birth_phases]
                fig.add_trace(
                    go.Scatter(x=birth_steps, y=[1]*len(births), mode='markers',
                              marker=dict(color=colors, size=10, symbol='star'),
                              name='Births', showlegend=True),
                    row=4, col=1
                )
            
            if deaths:
                death_steps, death_phases = zip(*deaths)
                colors = ['red' if p == 'A' else 'pink' for p in death_phases]
                fig.add_trace(
                    go.Scatter(x=death_steps, y=[0]*len(deaths), mode='markers',
                              marker=dict(color=colors, size=10, symbol='x'),
                              name='Deaths', showlegend=True),
                    row=4, col=1
                )
        
        # Adaptive thresholds
        if self.cfg.adaptive_thresholds and 'birth_threshold' in self.history:
            fig.add_trace(
                go.Scatter(x=self.history['step'], y=self.history['birth_threshold'],
                          name='Birth Threshold', line=dict(color='orange')),
                row=5, col=1
            )
            fig.add_trace(
                go.Scatter(x=self.history['step'], 
                          y=[t*100 for t in self.history['death_threshold']],
                          name='Death Threshold (×100)', line=dict(color='red', dash='dash')),
                row=5, col=1
            )
        
        fig.update_layout(
            template='plotly_dark',
            height=1200,
            title_text="Neuroplastic LSTM Training Progress"
        )
        
        fig.show()
        self.save_plot(fig, "training_progress")
    
    def plot_advanced_metrics(self):
        """Plot advanced metrics with save functionality"""
        # 1. Timeline stacked area chart
        plt.figure(figsize=(12, 6))
        
        births_cumsum = []
        deaths_cumsum = []
        steps = []
        births_count = 0
        deaths_count = 0
        
        for event in sorted(self.model.life_events, key=lambda x: x['step']):
            if event['type'] == 'birth':
                births_count += 1
            else:
                deaths_count += 1
            steps.append(event['step'])
            births_cumsum.append(births_count)
            deaths_cumsum.append(deaths_count)
        
        if steps:
            # Add phase markers
            phase_b_births = [e['step'] for e in self.model.life_events 
                            if e.get('phase') == 'B' and e['type'] == 'birth']
            if phase_b_births:
                plt.axvline(x=min(phase_b_births), color='yellow', linestyle='--', 
                          alpha=0.5, label='Phase B Start')
            
            plt.stackplot(steps, births_cumsum, deaths_cumsum, 
                         labels=['Births', 'Deaths'], 
                         colors=['#00ff88', '#ff0088'], 
                         alpha=0.7)
            plt.legend()
            plt.title("Cumulative Births and Deaths")
            plt.xlabel("Step")
            plt.ylabel("Count")
            self.save_matplotlib_plot("cumulative_births_deaths")
            plt.show()
        
        # 2. Fast weight vs contribution scatter
        if self.model.fw_norm_hist and self.model.contrib_hist:
            plt.figure(figsize=(16, 6))
            
            fw_norm = list(self.model.fw_norm_hist)[-1].numpy()
            contrib = list(self.model.contrib_hist)[-1].numpy()
            grad_contrib = self.model.grad_contrib.cpu().numpy()
            age = list(self.model.age_hist)[-1].numpy()
            mask = self.model.mask.cpu().numpy()
            
            active = mask > 0
            
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
            
            # Plot 1: FW norm vs contribution
            scatter1 = ax1.scatter(fw_norm[active], contrib[active], 
                                 c=age[active], cmap='viridis', 
                                 s=100, alpha=0.6, edgecolors='white')
            plt.colorbar(scatter1, label='Age', ax=ax1)
            ax1.set_xlabel("Fast Weight Norm")
            ax1.set_ylabel("Total Contribution")
            ax1.set_title("Fast Weight Learning vs Neuron Utility")
            ax1.set_yscale('log')
            ax1.grid(True, alpha=0.3)
            
            # Plot 2: Contribution components
            scatter2 = ax2.scatter(contrib[active], grad_contrib[active],
                                 c=age[active], cmap='plasma',
                                 s=100, alpha=0.6, edgecolors='white')
            plt.colorbar(scatter2, label='Age', ax=ax2)
            ax2.set_xlabel("Total Contribution")
            ax2.set_ylabel("Gradient Contribution")
            ax2.set_title("Contribution Components")
            ax2.set_xscale('log')
            ax2.set_yscale('log')
            ax2.grid(True, alpha=0.3)
            
            plt.tight_layout()
            self.save_matplotlib_plot("contribution_analysis")
            plt.show()
        
        # 3. Phase comparison
        if len([e for e in self.model.life_events if e.get('phase') == 'B']) > 0:
            self.plot_phase_comparison()
    
    def plot_phase_comparison(self):
        """Compare metrics between phases"""
        fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=('Birth Rate by Phase', 'Death Rate by Phase', 
                          'Average Lifespan', 'Contribution Distribution'),
            specs=[[{'type': 'bar'}, {'type': 'bar'}],
                  [{'type': 'bar'}, {'type': 'violin'}]]
        )
        
        # Separate events by phase
        phase_a_births = [e for e in self.model.life_events if e['type'] == 'birth' and e.get('phase', 'A') == 'A']
        phase_b_births = [e for e in self.model.life_events if e['type'] == 'birth' and e.get('phase', 'A') == 'B']
        phase_a_deaths = [e for e in self.model.life_events if e['type'] == 'death' and e.get('phase', 'A') == 'A']
        phase_b_deaths = [e for e in self.model.life_events if e['type'] == 'death' and e.get('phase', 'A') == 'B']
        
        # Birth rates
        fig.add_trace(
            go.Bar(x=['Phase A', 'Phase B'], 
                  y=[len(phase_a_births), len(phase_b_births)],
                  marker_color=['blue', 'lightblue']),
            row=1, col=1
        )
        
        # Death rates
        fig.add_trace(
            go.Bar(x=['Phase A', 'Phase B'], 
                  y=[len(phase_a_deaths), len(phase_b_deaths)],
                  marker_color=['red', 'pink']),
            row=1, col=2
        )
        
        fig.update_layout(template='plotly_dark', height=800, showlegend=False)
        fig.show()
        self.save_plot(fig, "phase_comparison")

    def plot_neuron_specialization(self):
        """Visualize neuron responses to prompts"""
        if not self.cfg.test_prompts:
            return
        
        device = next(self.model.parameters()).device
        for prompt in self.cfg.test_prompts:
            self.model.test_prompt(prompt, self.model.stoi, device)
        
        fig = go.Figure()
        
        prompt_data = []
        prompt_names = []
        
        for prompt in self.cfg.test_prompts:
            if prompt in self.model.prompt_activations:
                latest = self.model.prompt_activations[prompt][-1]
                prompt_data.append(latest['mean'])
                prompt_names.append(prompt[:30] + "..." if len(prompt) > 30 else prompt)
        
        if prompt_data:
            act_matrix = np.stack(prompt_data)
            active_mask = self.model.mask.cpu().numpy() > 0
            act_matrix = act_matrix[:, active_mask]
            
            fig.add_trace(
                go.Heatmap(
                    z=act_matrix,
                    x=[f"N{i}" for i in np.where(active_mask)[0]],
                    y=prompt_names,
                    colorscale='Viridis'
                )
            )
            
            fig.update_layout(
                title="Neuron Activation Patterns by Prompt",
                xaxis_title="Neuron ID",
                yaxis_title="Prompt",
                template='plotly_dark',
                height=400 + 20 * len(prompt_names)
            )
            
            fig.show()
            self.save_plot(fig, "neuron_specialization")
    
    def analyze_top_neurons(self, k=10):
        """Analyze top contributing neurons"""
        active_mask = self.model.mask > 0
        active_idx = active_mask.nonzero(as_tuple=True)[0]
        
        if len(active_idx) == 0:
            print("No active neurons!")
            return
        
        contribs = self.model.contrib[active_idx]
        top_idx = active_idx[contribs.argsort(descending=True)[:k]]
        
        print(f"\n🏆 Top {k} neurons by contribution:")
        print("-" * 80)
        
        neuron_data = []
        for rank, idx in enumerate(top_idx):
            info = self.model.get_neuron_info(idx.item())
            neuron_data.append(info)
            
            print(f"Rank {rank+1}: Neuron {idx.item()}")
            print(f"  Age: {info['age']} steps")
            print(f"  Total contribution: {info['contribution']:.6f}")
            print(f"  Gradient contribution: {info['grad_contribution']:.6f}")
            print(f"  Output impact: {info['output_impact']:.6f}")
            print(f"  Firing rate: {info['firing_rate']*100:.1f}%")
            print(f"  Fast weight norm: {info['fw_norm']:.3f}")
            
            # Birth/death history
            if info['birth_death']:
                birth_phase = [e['phase'] for e in info['birth_death'] if e['type'] == 'birth']
                if birth_phase:
                    print(f"  Born in phase: {birth_phase[0]}")
            print()
        
        return neuron_data
    
    def generate_report(self):
        """Generate comprehensive analysis report"""
        print("\n" + "="*80)
        print("NEUROPLASTICITY ANALYSIS REPORT")
        print("="*80)
        
        # Phase-separated statistics
        phase_a_births = len([e for e in self.model.life_events if e['type'] == 'birth' and e.get('phase', 'A') == 'A'])
        phase_b_births = len([e for e in self.model.life_events if e['type'] == 'birth' and e.get('phase', 'A') == 'B'])
        phase_a_deaths = len([e for e in self.model.life_events if e['type'] == 'death' and e.get('phase', 'A') == 'A'])
        phase_b_deaths = len([e for e in self.model.life_events if e['type'] == 'death' and e.get('phase', 'A') == 'B'])
        
        current_active = int(self.model.mask.sum().item())
        
        print(f"\n📊 Population Dynamics:")
        print(f"  Initial neurons: {self.cfg.initial_active}")
        print(f"  Current active: {current_active}")
        print(f"\n  Phase A:")
        print(f"    Births: {phase_a_births}")
        print(f"    Deaths: {phase_a_deaths}")
        print(f"  Phase B:")
        print(f"    Births: {phase_b_births}")
        print(f"    Deaths: {phase_b_deaths}")
        print(f"\n  Total births: {phase_a_births + phase_b_births}")
        print(f"  Total deaths: {phase_a_deaths + phase_b_deaths}")
        print(f"  Net change: {(phase_a_births + phase_b_births) - (phase_a_deaths + phase_b_deaths)}")
        
        if self.cfg.adaptive_thresholds:
            print(f"\n🎯 Adaptive Thresholds:")
            print(f"  Final birth threshold: {self.model.current_birth_threshold:.3f}")
            print(f"  Final death threshold: {self.model.current_death_threshold:.4f}")
        
        # Memory usage
        print(f"\n💾 Memory Efficiency:")
        print(f"  Sampling rate: {self.cfg.sampling_rate*100:.0f}%")
        print(f"  History buffer size: {self.cfg.neuron_history_size}")
        print(f"  Estimated memory saved: ~{(1-self.cfg.sampling_rate)*100:.0f}%")
        
        print("\n" + "="*80)
    
    def save_all_results(self):
        """Save all analysis results"""
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        
        # Save model checkpoint
        checkpoint_path = f"{self.cfg.checkpoints_dir}/plastic_lstm_{timestamp}.pt"
        torch.save({
            'model_state_dict': self.model.state_dict(),
            'life_events': self.model.life_events,
            'config': self.cfg,
            'history': dict(self.history),
            'timestamp': timestamp
        }, checkpoint_path)
        print(f"\n💾 Saved checkpoint: {checkpoint_path}")
        
        # Save metrics as JSON
        import json
        metrics_path = f"{self.cfg.checkpoints_dir}/metrics_{timestamp}.json"
        with open(metrics_path, 'w') as f:
            json.dump({
                'history': dict(self.history),
                'life_events_summary': {
                    'total_births': len([e for e in self.model.life_events if e['type'] == 'birth']),
                    'total_deaths': len([e for e in self.model.life_events if e['type'] == 'death']),
                    'final_active': int(self.model.mask.sum().item())
                },
                'timestamp': timestamp
            }, f, indent=2)
        print(f"💾 Saved metrics: {metrics_path}")

def train_plastic_lstm(cfg):
    """Main training loop with Phase B support"""
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")
    
    # Initial data (Phase A)
    train_loader, val_loader, stoi, itos, dataset = build_loaders(cfg, phase='A')
    
    # Model
    model = PlasticLSTM(cfg).to(device)
    print(f"Model initialized with {cfg.initial_active} active neurons")
    
    model.stoi = stoi
    model.itos = itos
    
    # Training setup
    optimizer = torch.optim.AdamW(model.parameters(), lr=cfg.learning_rate)
    replay_buffer = ReplayBuffer(cfg.replay_size)
    analyzer = NeuronAnalyzer(model, cfg, val_loader, itos)
    
    # Training loop
    step = 0
    model.train()
    phase_b_triggered = False
    
    for epoch in range(cfg.num_epochs):
        print(f"\n{'='*60}")
        print(f"Epoch {epoch+1}/{cfg.num_epochs}")
        print(f"{'='*60}")
        
        pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}')
        
        for batch_idx, (x, y) in enumerate(pbar):
            step += 1
            
            # PHASE B INJECTION
            if step >= cfg.phase_b_start and not phase_b_triggered:
                print(f"\n🚀 PHASE B: Injecting new data at step {step}!")
                model.set_phase('B')
                
                # Load new data while keeping vocabulary
                train_loader_b, val_loader_b, _, _, _ = build_loaders(
                    cfg, phase='B', existing_stoi=stoi, existing_itos=itos
                )
                
                # Switch to new data
                train_loader = train_loader_b
                val_loader = val_loader_b
                analyzer.val_loader = val_loader_b
                
                phase_b_triggered = True
                
                # Reset iterator
                pbar = tqdm(train_loader, desc=f'Epoch {epoch+1} - Phase B')
                x, y = next(iter(train_loader))
            
            x, y = x.to(device), y.to(device)
            
            # Forward pass
            optimizer.zero_grad()
            logits = model(x)
            loss = F.cross_entropy(logits, y[:, -1])
            
            # Backward pas
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            
            # Hebbian update
            model.hebbian_update()
            
            # Structural plasticity
            if step % cfg.plasticity_interval == 0:
                model.structural_update(loss.item())
            
            # Store in replay buffer
            replay_buffer.push(x, y)
            
            # Replay training
            if step % 100 == 0 and len(replay_buffer.buf) > 10:
                rx, ry = replay_buffer.sample()
                rx, ry = rx.to(device), ry.to(device)
                optimizer.zero_grad()
                r_logits = model(rx)
                r_loss = F.cross_entropy(r_logits, ry[:, -1])
                r_loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                optimizer.step()
            
            # Evaluation
            if step % cfg.eval_every_steps == 0:
                val_loss, val_ppl = analyzer.evaluate(val_loader, device)
                model.train()
                active_neurons = int(model.mask.sum().item())
                
                analyzer.log_metrics(step, val_loss, val_ppl, active_neurons, 
                                   phase='B' if phase_b_triggered else 'A')
                
                pbar.set_postfix({
                    'loss': f'{loss.item():.3f}',
                    'val_ppl': f'{val_ppl:.1f}',
                    'active': active_neurons,
                    'births': sum(1 for e in model.life_events if e['type'] == 'birth'),
                    'deaths': sum(1 for e in model.life_events if e['type'] == 'death'),
                    'phase': 'B' if phase_b_triggered else 'A'
                })
                
                # Test prompts
                if step % (cfg.eval_every_steps * 2) == 0:
                    print(f"\n📝 Testing prompts at step {step}:")
                    for prompt in cfg.test_prompts[:3]:
                        acts_mean, acts_std = model.test_prompt(prompt, stoi, device)
                        active_acts = acts_mean[model.mask.cpu().numpy() > 0]
                        print(f"  '{prompt[:20]}...': mean_act={active_acts.mean():.3f}, "
                              f"max_act={active_acts.max():.3f}")
        
        print(f"\nEpoch {epoch+1} complete!")
        print(f"Phase: {'B' if phase_b_triggered else 'A'}")
        print(f"Active neurons: {int(model.mask.sum().item())}/{cfg.max_neurons}")
    
    # Final analysissssss
    print("\n" + "="*60)
    print("TRAINING COMPLETE - FINAL ANALYSIS")
    print("="*60)
    
    # Test all prompts
    print("\n📊 Final prompt analysis:")
    for prompt in cfg.test_prompts:
        acts_mean, acts_std = model.test_prompt(prompt, stoi, device, num_runs=10)
        active_acts = acts_mean[model.mask.cpu().numpy() > 0]
        print(f"\n'{prompt}':")
        print(f"  Active neurons: {(active_acts > 0.1).sum()}/{len(active_acts)}")
        print(f"  Mean activation: {active_acts.mean():.3f} ± {acts_std[model.mask.cpu().numpy() > 0].mean():.3f}")
    
    # Generate all visualizationsssss
    print("\n📊 Generating visualizations...")
    analyzer.plot_training_progress()
    analyzer.plot_advanced_metrics()
    analyzer.plot_neuron_specialization()
    
    # Analysis
    analyzer.analyze_top_neurons()
    analyzer.generate_report()
    
# Save everythingss
    analyzer.save_all_results()
    
    print(f"\n✅ All results saved tooo {cfg.plots_dir} and {cfg.checkpoints_dir}")
    print("🎉 Training complete!")
    
    return model, analyzer

if __name__ == "__main__":
    model, analyzer = train_plastic_lstm(cfg)
```
